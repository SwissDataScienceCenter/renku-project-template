```python
T1 = "2022-01-01T00:00:00" # https://odahub.io/ontology#StartTimeISOT
T2 = "2022-01-03T00:00:00" # https://odahub.io/ontology#StopTimeISOT
n_bins = 100 # https://odahub.io/ontology#TimeBinSeconds
```

See https://odahub.io/docs/guide-development/ for more details about how to develop a workflow for use in Multi-Messenger ODA.

more detailed example in https://renkulab.io/projects/dsavchenko/mmoda-fink-backend

```python
import numpy as np
from astropy.time import Time
from oda_api.data_products import LightCurveDataProduct

lc = LightCurveDataProduct.from_arrays(
        times=Time(
            np.linspace(
                Time(T1, format='isot').mjd,
                Time(T2, format='isot').mjd,
                n_bins), 
            format="mjd"),
        fluxes=np.random.uniform(0, 1, n_bins),
        errors=np.random.uniform(0, 1, n_bins),
    )

obj_results = [lc.encode()]
```

```python
from matplotlib import pylab as plt

plt.figure()
plt.errorbar(
    lc.data_unit[1].data['TIME'],
    lc.data_unit[1].data['FLUX'],
    lc.data_unit[1].data['ERROR']
)
plt.ylabel('Flux')
plt.xlabel('Time, MJD')
```
